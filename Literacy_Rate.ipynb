### Which State has highest literacy Rate in 2001 ?

### Which State has lowest literacy Rate in 2001 ?

### Which State has highest literacy Rate in 2011 ?

### Which State has the highest Literacy Growth Rate in Rural Areas ?

### Which State has the highest Literacy Growth Rate in Urban Areas ?

### Graph showing the Literacy Rate in 2011.

```python
plt.figure(figsize=(12,6))
x=df1["State"].to_numpy()
y=df["Literacy Rate (Persons) - Total - 2011"].to_numpy()
plt.bar(x,y,color="c")
plt.xticks(x,rotation="vertical")
plt.title("Literacy Rate in 2011")
plt.ylabel("Literacy Rate")
plt.xlabel("States")
plt.show()
```

### Garph showing the Change of Literacy Rate from 2001 to 2011.

```python
plt.figure(figsize=(12,8))
style.use("ggplot")
x=df1["State"].to_numpy()
y=df["Literacy Rate (Persons) - Total - 2001"].to_numpy()
y1=df["Literacy Rate (Persons) - Total - 2011"].to_numpy()
plt.plot(x,y,marker="*",ms=10,mfc="r",mec="b",label="2001",linewidth=2)
plt.plot(x,y1,marker="*",ms=10,mfc="y",mec="g",label="2011",linewidth=2)
plt.title("Literacy Rate")
plt.ylabel("Literacy")
plt.xlabel("States")
plt.xticks(x,rotation="vertical")
plt.legend()
plt.grid(True,color="w")
plt.show()
```

### Literacy Rate in Rural and Urban Areas in 2001.

```python
plt.figure(figsize=(12,6))
LRD_2001=df["Literacy Rate (Persons) - Urban - 2001"]-df["Literacy Rate (Persons) - Rural - 2001"]
x=df1["State"].to_numpy()
y=df["Literacy Rate (Persons) - Rural - 2001"].to_numpy()
y1=LRD_2001.to_numpy()

plt.plot([],[],color="g",label="Rural-2001",linewidth=5)
plt.plot([],[],color="m",label="Urban-2001",linewidth=5)

plt.stackplot(x,y,y1,colors=["g","m"])

plt.ylabel("Literacy Rate")
plt.xlabel("State")
plt.title("Comparison of Literacy Rate in Rural and Urban Areas in 2001")
plt.xticks(x,rotation="vertical")
plt.legend()
plt.show()
```

### Literacy Rate in Rural and Urban Areas in 2011.

```python
plt.figure(figsize=(12,6))
LRD_2011=df["Literacy Rate (Persons) - Urban - 2011"]-df["Literacy Rate (Persons) - Rural - 2011"]
x=df1["State"].to_numpy()
y=df["Literacy Rate (Persons) - Rural - 2011"].to_numpy()
y1=LRD_2011.to_numpy()

plt.plot([],[],color="r",label="Rural-2011",linewidth=5)
plt.plot([],[],color="orange",label="Urban-2011",linewidth=5)

plt.stackplot(x,y,y1,colors=["r","orange"])

plt.ylabel("Literacy Rate")
plt.xlabel("State")
plt.title("Comparison of Literacy Rate in Rural and Urban Regions in 2011")
plt.xticks(x,rotation="vertical")
plt.legend()
plt.show()
```

### Literacy Rate of 2001 & 2011 in Rural Areas.

```python
plt.figure(figsize=(12,8))
style.use("ggplot")
x=df1["State"].to_numpy()
y=df["Literacy Rate (Persons) - Rural - 2011"].to_numpy()
y1=df["Literacy Rate (Persons) - Rural - 2001"].to_numpy()
plt.plot(x,y,"m",marker="*",mec="r",label="Rural_2011",linewidth=2)
plt.plot(x,y1,"b",marker="*",mec="g",label="Rural_2001",linewidth=2)
plt.title("Literacy Rate of 2001 & 2011 in Rural Areas")
plt.ylabel("Literacy Rate")
plt.xlabel("State")
plt.legend()
plt.xticks(x,rotation="vertical")
plt.grid(True,color="w")
plt.show()
```

### Literacy Rate of 2001 & 2011 in Urban Areas.

```python
plt.figure(figsize=(12,8))
style.use("ggplot")
x=df1["State"].to_numpy()
y=df["Literacy Rate (Persons) - Urban - 2011"].to_numpy()
y1=df["Literacy Rate (Persons) - Urban - 2001"].to_numpy()
plt.plot(x,y,"c",marker="*",mec="b",label="Rural_2011",linewidth=2)
plt.plot(x,y1,"g",marker="*",mec="r",label="Rural_2001",linewidth=2)
plt.title("Literacy Rate of 2001 & 2011 in Urban Areas")
plt.ylabel("Literacy Rate")
plt.xlabel("State")
plt.legend()
plt.xticks(x,rotation="vertical")
plt.grid(True,color="w")
plt.show()
```
